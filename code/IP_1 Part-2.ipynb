# D1 for USA

```python
D1=new_frame.loc[(new_frame['country_code']=="USA") & (new_frame['raised_amount_usd']>=5000000) & (new_frame['raised_amount_usd']<=15000000)]
D1.head()
```

```python
#total number of investments for USA
usa=pd.DataFrame(D1.groupby('Main_Sector')['raised_amount_usd'].count().sort_values(ascending=False))
usa.rename(columns={'raised_amount_usd':'Total_no_of_investment'},inplace=True)
usa
```

```python
#total amount of investments for USA
usa_total=pd.DataFrame(D1.groupby('Main_Sector')['raised_amount_usd'].sum().sort_values(ascending=False).apply(lambda x: '{:.2f}'.format(x)))
usa_total.rename(columns={'raised_amount_usd':'Total_of_investment'},inplace=True)
usa_total
```

```python
#merging the columns
D1=pd.merge(D1,usa,on='Main_Sector')
D1=pd.merge(D1,usa_total,on="Main_Sector")
```

```python
D1.columns
```

```python
D1
```

```python
D1.shape
```

# D2 for GBR

```python
D2=new_frame.loc[(new_frame['country_code']=="GBR") & (new_frame['raised_amount_usd']>=5000000) & (new_frame['raised_amount_usd']<=15000000)]
D2
```

```python
#total number of investments for GBR
gbr=pd.DataFrame(D2.groupby('Main_Sector')['raised_amount_usd'].count().sort_values(ascending=False))
gbr.rename(columns={'raised_amount_usd':'Total_no_of_investment'},inplace=True)
gbr
```

```python
#total amount of investments for GBR
gbr_total=pd.DataFrame(D2.groupby('Main_Sector')['raised_amount_usd'].sum().sort_values(ascending=False))
gbr_total.rename(columns={'raised_amount_usd':'Total_of_investment'},inplace=True)
gbr_total
```

```python
#merging the columns
D2=pd.merge(D2,gbr,on='Main_Sector')
D2=pd.merge(D2,gbr_total,on="Main_Sector")
```
